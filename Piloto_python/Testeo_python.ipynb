```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

# Ejemplo de uso:
arr = [5, 2, 8, 1, 4]
bubble_sort(arr)
print(arr)  # Imprime [1, 2, 4, 5, 8]
```

# Tarea N2

Desarrollo de un algortimo que permita la suma de manera inversa

    Se deduce que "inverso" hacer referencia a sumar los valores comenzando desde el ultimo hasta el primero

```python
def sum_inversa(A: list[float]):
    n=len(A)
    sum = 0
    for i in range (n, 0,-1):
        sum = sum + A[i-1]
        print(sum)
        i-=1
    print(sum)
```

```python
A=[1, 5, 7, 8]
sum_inversa(A)
A
```

```python
B=[1, 4, 7, 8, 5, 7, 8]
sum_inversa(B)
```

Mostar particular atencion a las tabulaciones que se apliquen en el codigo!

Asi como tambien, condiderar, el nombre de las variables adecuadadas

```python
def sort_select(A: list[float]):
    n=len(A)
    for i in range(1, n-1):
        min_index=i
        for j in range(i+1, n):
            if A[j]<A[min_index]:
                min_index = j
        temp=A[i]
        A[i]=A[min_index]
        A[min_index]=temp
        print(A)
```

```python
A=[1, 7, 3, 5]
sort_select(A)
```
