```python
import matplotlib.pyplot as plt
import numpy as np

# Definimos la funci√≥n
def f(x):
  return x**2 - 6*x + 5

# Calculamos las ra√≠ces de la ecuaci√≥n (usando la f√≥rmula general)
a = 1
b = -6
c = 5
discriminante = b**2 - 4*a*c
if discriminante >= 0:
  x1 = (-b + np.sqrt(discriminante)) / (2*a)
  x2 = (-b - np.sqrt(discriminante)) / (2*a)
  raices = [x1, x2]
else:
  print("La ecuaci√≥n no tiene ra√≠ces reales.")

# Creamos un rango de valores para x
x = np.linspace(-2, 8, 100)

# Evaluamos la funci√≥n en los valores de x
y = f(x)

# Creamos el gr√°fico
plt.plot(x, y)
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Gr√°fica de y = x^2 - 6x + 5')
plt.grid(True)

# Marcamos las ra√≠ces
if discriminante >= 0:
  plt.scatter(raices, [0]*len(raices), color='red', marker='o')
  #plt.scatter(raices, [0]*len(raices), label="")


# Mostramos el gr√°fico
plt.show()
```

```python
def biseccion(f, a, b, tol=1e-6):
    """
    Encuentra una ra√≠z de la funci√≥n f en el intervalo [a, b] usando el m√©todo de bisecci√≥n.

    Args:
        f: La funci√≥n cuya ra√≠z se busca.
        a: Extremo izquierdo del intervalo.
        b: Extremo derecho del intervalo.
        tol: Tolerancia (error).

    Returns:
        La ra√≠z aproximada o None si no se encuentra una ra√≠z.
    """

    if f(a) * f(b) >= 0:
        print("La funci√≥n no cambia de signo en el intervalo dado.")
        return None

    while (b - a) >= tol:
        c = (a + b) / 2
        if f(c) == 0.0:
            break
        if f(c) * f(a) < 0:
            b = c
        else:
            a = c
            print(c)

    return c

# Ejemplo de uso:
f = lambda x: x**2 - 6*x +5

root = biseccion(f, -2, 3)
print("La ra√≠z aproximada es:", root)

root = biseccion(f, 3, 8)
print("La ra√≠z aproximada es:", root)
```

# **Metodo de la Biseccion**

```python
def own_biseccion(f, a, b, tol, N):
    FA=f(a)
    i=1
    while i <= N:
        p=a+((b-a)/2)
        FP= f(p)
        if FP < tol:
            if FP == 0.0:
                break
            elif FP == (b-a)/2:
                return p
        i+=1
        if FA*FP > 0:
            a=p
            return p
        elif FA == FP:
            b=p
            return p
    return p
```

```python
f = lambda x: x**2 - 6*x +5

root = own_biseccion(f, -2, 3, 1e-6, 20)
print("La ra√≠z aproximada es:", root)

root = own_biseccion(f, 3, 4, 1e-2, 5)
print("La ra√≠z aproximada es:", root)
```

# **Metodo de la Secante**

Los siguientes ejemplos estan basados en el uso de la liberia scipy

Dependiendo de los valores que agregemos en los parametros, la funcion cambiara de una u otra manera

Los ejercicios de acontinuacion estan basados en el **Metodo de la Secante**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

f = lambda x: x**3 - 3*x**2 + x -1 

root = optimize.newton(f, 100)
root
```

```python
root = optimize.newton(f, 1)
root
```

```python
# La funcion no converge despues de 50 iteraciones y se crea un error

root = optimize.newton(f, 1) # valor inscrito de prueba (f, 0.75)
root
```

```python
root = optimize.newton(f, 1.5, fprime=lambda x: 3 * x**2)
root
```

    - No necesita derivadas
    - Se necesitan dos estimaciones inciales 
    - Mas rapida que Biseccion
    - Inestables 
    - Es iterativa

# **Metodo de la Newton**

    - Necesita la derivada de la funcion para encontrar el valor de convergencia
    - Si la derivada da 0, entonces la convergencia se vuelver indefinida
    - Requiere de una estimacion inicial
    - A la larga se vuelve inestable y genera divergencia
    - Tambien es iterativo
     

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

f = lambda x: x**3 - 3*x**2 + x -1 

root = optimize.newton(f, 100, fprime=lambda x: 3 * x**2 - 6*x + 1)
root
```
